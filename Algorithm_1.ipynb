```python
1. 알고리즘이란....
- 문제나 과제를 해결하기 위한 처리 절차를 구체적인 순서로 표현한 아이디어 또는 생각
- 알고리즘 자체는 일상생활에서도 많이 이미 사용하고 있다.
- 요리의 레시피, 음악의 악보, 가전제품 사용설명서 등
- 문자, 기호, 일러스트 등으로 나타낸다.

2. 알고리즘과 프로그램
- 알고리즘을 프로그래밍 언어로 기술하면 프로그램이 된다.
- 프로그램밍 언어(파이썬, 자바, 씨...)는 컴퓨터에 지시하기 위한 인공 언어이다.

3. 프로그램 작성에서의 알고리즘
- 프로그램 작성은 기획, 설계(*), 프로그래밍, 디버그로 진행한다.
- 알고리즘은 설계 단계에서 필요하다.

4. 좋은 알고리즘이란??
- 알기 쉽다. (쉬운 이해)
- 속도가 빠르다.
- 효율적이다. (메모리를 작게 사용한다.)
- 재이용이 쉽다. (객체지향프로그램)

5. 왜 알고리즘을 공부해야 하나
- 좋은 프로그램을 만들기위해 (빠르고, 효율적, 범용성)
- 프로그램의 좋고 나쁨을 판단하기 위해
- 프로그램 작성 과정 전체를 효율화하기 위해
- 프로그램 기술을 향상시키기 위해

6. 알고리즘의 조건
- 정확한 결과를 얻을 수 있어야 한다.
  (알고리즘은 문제나 과제를 해결하기 위한 절차)
- 반드시 종료되어야 한다.
  (중간 처리나 절차가 잘못 만들어지면 영원히 끝나지 않는다.)

7. 알고리즘의 기본형 3가지
- 순차 구조: 처음부터 순서대로 처리하는 절차
  예) 심부름
      1 슈퍼마켓에 간다.
      2 라면을 산다.
      3 집으로 돌아온다.
  실행을 원하는 처리를 위로부터 순서대로 작성한다.

- 선택 구조 : 조건식으로 판단해 실핼할 절차를 전환하는 구조
```

```python
#반복문으로 팩토리얼 구하기

#함수 선언
def facto(n):
 #변수 선언
 output = 1
 #반복문으로 곱하기
 for i in range(1,n+1)
  output *= i
 return output

print('3!:', facto(3))
print('3!:', facto(5))
print('3!:', facto(7))
```

```python
# 재귀(recursion)로 팩토리얼 구하기
def facto(n):
 if n == 0:
   return 1
  else:
   return n * facto(n-1)

print('3!:', facto(3))
print('3!:', facto(5))
print('3!:', facto(7))

# 개발자들 사이에는 재귀함수를 절대 사용하지 말자 라는 의견도 있다.
# 하지만 가독성이 좋고 코드가 간결하고 멋스러워 보이는 장점이 있다.(예쁜 쓰레기)
```

```python
try:
# 예외가 발생할 가능성이 있는 코드

except:
# 예외가 발생했을 때 실행할 코드

else:
# 예외가 발생하지 않았을 때 실행할 코드

finally:
# 항상 실행 시킬 코드
```

```python
try:
  num = int(input('정수를 입력하세요>'))
except:
  print('정수를 입력하지 않았습니다.')
else:
  print('입력한 정수는',num,'입니다')
finally:
  print('프로그램이 완전히 종료되었습니다.')
```

```python
def test():
  print('test()함수의 첫 줄입니다.')
  try:
    print('try 구문이 실행되었습니다.')
    return
    print('try 구문의 return 아래입니다.')
  except:
    print('except구문이 실행되었습니다.')
  else:
    print('else구문이 실행되었습니다.')
  finally:
    print('finally 구문이 실행되었습니다.')
  print('test() 함수의 마지막 줄입니다.')

test()
```

```python
print('프로그램 실행')

while True:
  try:
    print('try 구문 실행')
    break
    print('try 구문의 break 아래')
  except:
    print('except 구문 실행')
  finally:
    print('finally 구문 실행')
  print('while의 마지막 줄')
print('프로그램 종료')
```
